```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv(r"C:\Users\offic\Desktop\cake\adventure_works_purchase_orders.csv")
```

```python
#load columns  to see data structure
print(data.columns)
print(data.head())
```

```python
#strip spaces 
data.columns = data.columns.str.strip()
```

```python
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
```

```python
data.set_index('OrderDate', inplace=True)
```

```python
data.sort_index(inplace=True)
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.plot(data.index, data['Unit_Cost'], label='Unit Cost', color='red', alpha=0.6)
```

```python
plt.plot(data.index, data['UnitPrice'], label='Unit Price', color='green', alpha=0.7)
```

```python
data.to_csv(r"C:\Users\offic\Desktop\cake\cleaned data.csv")
```
